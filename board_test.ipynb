```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Take input from user
# n = int(input("Enter number of data points: "))
X = [1,2,3,4,5,6,7,8]
y = [2,5,6,9,10,10,15,20]
# for i in range(n):
#     x_val = float(input(f"Enter X[{i+1}]: "))
#     y_val = float(input(f"Enter y[{i+1}]: "))
#     X.append(x_val)
#     y.append(y_val)

X = np.array(X).reshape(-1, 1)
y = np.array(y)

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Take a new input for prediction
x_pred = float(input("Enter a value to predict y: "))
y_pred = model.predict([[x_pred]])
print(f"Predicted y value: {y_pred[0]}")

# Plot the data and regression line
# plt.scatter(X, y, color='blue', label='Data Points')
# plt.plot(X, model.predict(X), color='red', label='Regression Line')
# plt.scatter([x_pred], [y_pred], color='green', marker='o', label='Prediction')
# plt.xlabel("X")
# plt.ylabel("y")
# plt.legend()
# plt.show()
```
